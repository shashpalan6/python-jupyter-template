### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
try:
    import matplotlib.pyplot as plt
except ImportError as e:
    print(f"Error importing matplotlib: {e}")
    raise  # Raise the error again to halt execution

# Data for the plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title('Bar Plot with Custom Colors')

# Show the plot
plt.show()
```
